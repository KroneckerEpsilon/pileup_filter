# Pile-up algorithm

## Data Conversion

Each event is in the end stored in a column the variables ```si1``` and ```si2```

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
d1 = "sipm1data0.csv"
d2 = "sipm2data0.csv"

si1 = np.loadtxt(d1, delimiter = ",")
si2 = np.loadtxt(d2, delimiter = ",")
tbase = np.loadtxt("tbase.csv", delimiter = ",")

ev = 4
````

````python
plt.figure(figsize = (10, 5))
plt.subplot(121)
plt.plot(tbase, si1[ev], label = "SiPM 1 Hit Nr " + str(ev))
plt.ylabel("Voltage [V]")
plt.xlabel("Time [s]")
plt.legend()
plt.subplot(122)
plt.plot(si2[ev], label = "SiPM 2 Hit Nr " + str(ev))
plt.ylabel("Volts")
plt.legend()
plt.tight_layout()
plt.show()
````

````python
print("Digital Sampling Time Constant")
print(tbase[0], tbase[1])
print("1 ADC corresponds to")
print(tbase[36800])
print(si1[4][36800])
print(tbase[36840])
print(si1[4][36840])
````

````python
si1_dig = np.around(si1/0.0008)
si2_dig = np.around(si2/0.0008)
np.savetxt(d1[:-4]+"_ev"+str(ev)+"_dig.csv", si1_dig[ev], fmt="%d", delimiter = "\n")
np.savetxt(d2[:-4]+"_ev"+str(ev)+"_dig.txt", si2_dig[ev], fmt="%d", delimiter = "\n")
````
